```
import numpy as np
import matplotlib.pyplot as plt
```

```
CIFR_Acc_MADAM = np.load('CIFR_Acc_MADAM.npy')
CIFR_Acc_SGD = np.load('CIFR_Acc_SGD.npy')
CIFR_Acc_ADAM = np.load('CIFR_Acc_ADAM.npy')

FMNIST_Acc_MADAM = np.load('FMNIST_Acc_MADAM.npy')
FMNIST_Acc_SGD = np.load('FMNIST_Acc_SGD.npy')
FMNIST_Acc_ADAM = np.load('FMNIST_Acc_ADAM.npy')
```

```
plt.style.use("seaborn")
plt.xlabel("Learning rate")
plt.ylabel("Accuracy")
plt.title("MADAM")
plt.xticks([0,1,2,3,4], ['$10^0$','$10^{-1}$','$10^{-2}$','$10^{-3}$','$10^{-4}$'])

temp = FMNIST_Acc_MADAM
y = (np.max(temp,axis=1)/100).tolist()
p = max(y)
q = y.index(p)
plt.scatter(q,p,marker='x',c='r',s=100,zorder=10,linewidth=3)

x = np.arange(len(y))
plt.plot(x,y,label = "Fashion MNIST")

temp = CIFR_Acc_MADAM
y = (np.max(temp,axis=1)/100).tolist()
p = max(y)
q = y.index(p)
plt.scatter(q,p,marker='x',c='r',s=100,zorder=10,linewidth=3)

x = np.arange(len(y))
plt.plot(x,y,label = "CIFAR10")

plt.legend()


plt.savefig('./Comparision_MADAM.png')
plt.show()

```
